```
import numpy as np
import matplotlib.pyplot as plt
```

```
input = np.linspace(-100, 65, 400)
```

```
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```
output = sigmoid(input)
```

```
output
```

```
plt.plot(input, output)
plt.xlabel("input values")
plt.ylabel("output values")

plt.title("sigmoid")
```
